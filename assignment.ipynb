Write a program that accepts student names and their marks in 3 subjects. Store this data in a dictionary where the key is the student name and the value is a tuple of marks.
Then,

Display each student’s average marks.

Display the student(s) with the highest average.

Write a function get_top_students(data_dict) that returns a list of students who have the highest average marks.

```python
data_dict={}
number_students=int(input("Enter number of students:"))

for i in range(number_students):
    name= input(f"Enter name of student number {i+1}:")
    marks1=int(input("Enter mark of the first subject"))
    marks2=int(input("Enter mark of the second subject"))
    marks3=int(input("Enter mark of the third subject"))
    
    #Storing data in a dictionary
    data_dict[name]=(marks1,marks2,marks3)
print(data_dict)

#Average marks
average= {student: sum(marks)/len(marks) for student, marks in data_dict.items()}
highest_avg= max(average.values())

top_student=[student for student, avg in average.items() if avg == highest_avg]
print(f"Top student is {top_student} with {highest_avg}")

def get_top_student(data_dict):
    average= {student: sum(marks)/len(marks) for student, marks in data_dict.items()}
    highest_avg= max(average.values())
    return [student for student, avg in average.items() if avg == highest_avg]



```
