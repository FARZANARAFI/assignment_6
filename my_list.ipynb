```python
#square value for even and cube for odd
even_squared=[x**2 if x%2==0 else x**3 for x in range (1,11)]
print(even_squared)
```

```python
matrix=[[1,2,3],[4,5,6],[7,8,9]]
flattened=[num for row in matrix for num in row]
print(flattened)
```

```python
def square_if_even(num):
    if num % 2 == 0:
        return num ** 2
    else:
        return num
    
squared_list=[square_if_even(x) for x in range(1,11)]
print(squared_list)
```
