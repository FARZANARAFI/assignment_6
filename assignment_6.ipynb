```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

#synthetic dataset
no_of_students=100

#studentid
student_ids=np.arange(1,no_of_students+1)

#marks of 5 subjects
marks = np.random.randint(30, 101, size=(no_of_students,5))


#creating dataframe

df=pd.DataFrame(np.column_stack((student_ids,marks)),
columns=["student_id","Maths","Science","English","History","Computer"])

#adding average column

df['Average'] = df[['Maths', 'Science', 'English','History','Computer']].mean(axis=1)

#grade 

def grades(avg):
    if avg>=85:
        return 'A'
    elif avg>=70:
        return 'B'
    elif avg>=50:
        return 'C'
    else :
        return 'D'


#adding grade column

df['Grade'] = df['Average'].apply(grades)
print(df)
    
```

```python
import os
#create output directory
os.makedirs("output/plots",exist_ok=True)


#Bar chart
# Show the plot
sns.countplot(data=df, x="Grade")

# Adding labels and a legend
plt.xlabel('Grade')
plt.ylabel('No of Students')
plt.title('Distribution of Grades')
plt.savefig("output/plots/grade_distribution.png")
plt.close()

# Show the plot
plt.show()


#Boxplot
Scores_of_subjects=["Maths","Science","English","History","Computer"]
sns.boxplot(data=df[Scores_of_subjects])

# Adding labels and a legend
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.title('Boxplot of scores for all subjects')
plt.savefig("output/plots/subject_boxplot.png")
plt.close()

# Show the plot
plt.show()


#Correlation heatmap

corr_matrix = df[Scores_of_subjects].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Subjects")
plt.savefig("output/plots/correlation_heatmap.png")
plt.close()
plt.show()

df.describe().to_csv("output/summary_statistics.csv",index=True)
```
