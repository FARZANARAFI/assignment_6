Dictionary methods
1.Add a key value pair.
2.Updating a dictionary with another dictionary.
3.remove=pop,


```python
student_info={'name':'Alice','age':23,'courses':['Math','Compsci']}
student_info['city']='Newyork'
print(student_info)
```

```python
additional_dict={'marks':85,'graduation':False}
student_info.update(additional_dict)
print(additional_dict)
print(student_info)
```

```python
removed_age=student_info.pop('age')
print("Removed age:",removed_age)
print(student_info)
```

```python

removed_pair=student_info.popitem()
print("Rmoved pair:",removed_pair)
print(student_info)
```

```python
del student_info['age']
print(student_info)
```

```python
#membership


print('name' in student_info)#True
print('salary'in student_info)#False
```

```python
keys=student_info.keys() #get all keys
values=student_info.values() #get all values
items=student_info.items() #get all items
print(keys)
print(values)
print(items)
```

```python
#Dictionary manipulation methods
#sorting a dictionary by keys

student_info={'name':'Alice','age':23,'courses':['Math','Compsci']}
sorted_by_keys=dict(sorted(student_info.items()))
print("Sorted by keys:",sorted_by_keys)
```

```python
#copying adictionary
student_info={'name':'Alice','age':23,'courses':['Math','Compsci']}


student_info_copy=student_info.copy()
print("Original:",student_info)
print("Copy:",student_info_copy)
```

```python
#Nested Dictionary

nested_dict={'student1':{'name':'Alice','age':23},'student2':{'name':'Bob','age':25},
'teacher_name':'Mr.Smith'}
print(nested_dict)

```

```python
name=nested_dict.get('student2')
print(name)
```

```python
nested_student2_name=nested_dict['student2']['name']
print(nested_student2_name)
```

```python
nested_student2_name=nested_dict.get('student2').get('name')
print(nested_student2_name)
```

Dictionary Comprehension

```python
square_numbers={1:1,2:4,3:9,4:16,5:25,6:36,7:49,8:64,8:72,9:81,10:100}
print(square_numbers)
```

```python
square_dict={i:i*i for i in range(5)}
print(square_dict)
```

```python
square_dict={i:i**0.5 for i in range(5)}
print(square_dict)
```

```python
words=['apple','banana','cherry']
word_length_dict={word:len(word) for word in words}
print(word_length_dict)
```

```python
keys=['name','age','city']
values=['Alice',23,'Nweyork']
info={keys[i]:values[i] for i in range(len(keys))}
print(info)
```
