```python
 #Set operations

#union

set_a={1,2,3}
set_b={3,4,5}
set_union=set_a.union(set_b)
print("union:",set_union)

set_union2=set_a | set_b
print("union using |:",set_union)
```

```python
#intersection

set_intersection=set_a.intersection(set_b)
print("intersection:",set_intersection)

set_intersection2=set_a & set_b
print("intersection using &:",set_intersection2)
```

```python
#differece

set_difference=set_a.difference(set_b)
print("difference :",set_difference)


set_difference2=set_a - set_b
print("difference2 using (A-B):",set_difference2)
```

```python
#Symmetric difference

set_sym_diff=set_a.symmetric_difference(set_b)
print("symmetric diff using :",set_sym_diff)

set_sym_diff2=set_a ^ set_b
print("symmetric diff using ^:",set_sym_diff2)
```
